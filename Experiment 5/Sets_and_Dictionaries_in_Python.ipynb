```python
'''Problem Statement 2: Three students choose elective subjects.
Find the common subjects chosen by all students using set operations.
Concept Used: Set intersection (&)'''
```

```python
Ram_electives = {"Math", "Physics", "Chemistry", "Biology"}
Rahul_electives = {"Physics", "Computer Science", "Math", "English"}
Rohan_electives = {"Math", "Biology", "Computer Science", "Physics"}

common_electives =Ram_electives & Rahul_electives & Rohan_electives

print("Subjects chosen by Ram:", Ram_electives)
print("Subjects chosen by Rahul:", Rahul_electives)
print("Subjects chosen by Rohan:", Rohan_electives)
print("\nCommon subjects chosen by all students:", common_electives)
```

```python
'''Problem Statement 4: Absent Students
Given a set of all students and a set of present students
Find the absent students.
Concept used: Difference'''
all_students = {"Alex" , "Joseph" , "John" , "Thomas" , "Tony" , "Calvin" , "Donald"}
present_students = {"Alex" , "Joseph" , "John" , "Thomas"}
absent_students = all_students - present_students
print("Absent students:", absent_students)
```

```python
'''2.A dictionary stores students name and marks.
Write a python program to search for a student's marks using the name and marks.
If the student is not found, display a suitable message.'''

marks = {"Arnav": 80, "Aryan": 85, "Rahul": 90}
name = input("Enter student name: ")
print(marks.get(name, "Student not found"))
```

```python
'''3.A dictionary stores usernames and passwords.
Write a Python program to validate user login.'''

users = {"admin": 'user@123'}
user_name = input("Enter username: ")
password = input("Enter password: ")
if users.get(user_name) == password:
    print("Login successful")
else:
    print("Invalid username or password")
```

```python
'''4.A dictionary stores student names and marks.
Write a Python program to find the student who scored the highest marks.'''

marks = {"Arnav": 80, "Aryan": 85, "Rahul": 90}
topper = max(marks, key=marks.get)
print("Topper:", topper)
print("Marks:", marks [topper])
```
