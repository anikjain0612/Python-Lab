```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignment_Completed': [3, 4, 3, 5, 6]
}
df = pd.DataFrame(data)

# Advanced Bar Chart with spacing
plt.figure(figsize=(10,6)) # Increased figure size for better visibility

bar_width = 0.35 # Width of the bars
index = range(len(df['Days'])) # Numeric index for positioning bars

plt.bar([i - bar_width/2 for i in index], df['Study_Hours'], bar_width, color='green', label='Study Hours')
plt.bar([i + bar_width/2 for i in index], df['Marks'], bar_width, color='red', label='Marks')

plt.title("Study Hours and Marks Comparison")
plt.xlabel("Days")
plt.ylabel("Value")
plt.xticks(index, df['Days']) # Set x-axis ticks to display Days
plt.legend()
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#2. Bar Chart (Comparison)
plt.bar(df['Days'], df['Marks'], color='pink')

plt.title("Marks Comparison")
plt.xlabel("Days")
plt.ylabel("Marks")

plt.show()
```

```python
plt.bar(df['Study_Hours'],df['Marks'],color = 'magenta',edgecolor = 'black',width=0.3)
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Days'], df['Marks'], color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')

plt.title("Marks_Comparison")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(True)
plt.grid(axis='y')
```

```python
#3.Histogram (Distribution)
#Shows how values aqre spread

data = {
    'Marks': [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]
}
df = pd.DataFrame(data)
plt.hist(df['Marks'], bins=3, edgecolor='black', alpha=1, color='aqua')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#4. Scatter Plot
import matplotlib.pyplot as plt
import pandas as pd # Ensure pandas is imported

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignment_Completed': [3, 4, 3, 5, 6]
}
# Create a new DataFrame 'df' from the 'data' dictionary
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'], df['Marks'])

plt.title("Study of Hours Vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
# Conditional scatter plot
#Example category
df['Result'] = ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']
colors = ['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c=colors)

plt.title("Study of Hours Vs Marks (Result-wise)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```
