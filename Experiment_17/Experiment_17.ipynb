```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from numpy.random.mtrand import random
np.random.seed(1)

df = pd.DataFrame ({
    'Category' : ['A','B','C','D','E'],
    'Values': [0,70,30,60,90],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(10,100,5),
})
df
```

```python
# 1. Area Plot

plt.figure()
plt.fill_between(df['Category'],df['Values'],color='olivedrab',alpha=0.35)
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid") # Apply Seaborn style

plt.figure(figsize=(8,6))

plt.fill_between(df['Category'],df['Sales'], color = 'lemonchiffon', alpha = 0.9, label ='Sales')
plt.fill_between(df['Category'],df['Profit'], color = 'darkorchid', alpha = 0.5, label ='Profit')

plt.title('Sales and Profit Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# 2. Pie Chart

plt.figure()
plt.pie(df['Values'], labels = df['Category'], autopct = '%1.2f%%')
plt.title('Pie Chart of Values by Category')
plt.show()
```

```python
plt.figure()
plt.pie(df['Values'], labels = df['Category'], autopct = '%1.2f%%', explode = (0,0,.1,0,0.3))
plt.title('Pie Chart of Values by Category')
plt.show()
```

```python
# 3. Donut Chart

from matplotlib.patches import Circle

plt.figure()
plt.pie(df['Values'], labels = df['Category'], autopct = '%1.1f%%')
centre_circle = Circle((0,0),0.70,fc='mistyrose')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart of Values by Category')
plt.show()
```

```python
#4. Boxplot (Outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'],)
plt.title("Boxplot of Sales")
plt.show()
```

```python
#5. Heatmap (Correlation)
plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Values','Sales','Profit']].corr()

plt.figure(figsize=(5,4))

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Bubble Plot
import matplotlib.pyplot as plt

plt.scatter(df['Sales'], df['Profit'],
            s=df['Values']*10)   # size of bubbles

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(50, 300),      # min-max size
                palette='viridis')

plt.title("Bubble Plot using Seaborn")

plt.show()
```

```python
#2. Second Dataset
np.random.seed(7)
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())
```

```python
#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("Loan Amount Distribution (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#3. Histogram (Credit Score Distribution)
plt.hist(df['Credit_Score'], bins=6, color='skyblue', edgecolor='black')

plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.show()
```

```python
#Bubble Plot (Advanced Scatter)
plt.scatter(df['Income'], df['Loan_Amount'],
            s=df['Credit_Score']/2,
            c=df['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("Bubble Plot (Income vs Loan)")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
